***Mode = Mode repeated number funds.***

***Find Mode without using library.***

```python
# Ques_01 find the mode of number without using library.
# give to value in list
number = [20,45,25,20,34,46,12,20,55,25,20,30,20]
freq = {} # empty dictionary
for x in number:
    freq[x] = freq.get(x,0)+1
    max_freq = max(freq.values())
    modes = []
for k,v in freq.items():
    if v == max_freq:
        modes.append(k)
print(modes)
```

***Find Mode with using library.***

***Import Library***

```python
import statistics # import library
# give to dataset
dataset = [20,45,10,25,10,22,10,44,11,33,10]
mode_val = statistics.mode(dataset)
print("Mode =",mode_val)
```
