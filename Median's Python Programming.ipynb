***Median = The Median value in a dataset that has been arranged in order from lowest to highest or highest to lowest.***

***Find Median without using library***

```python
# Ques_01 find the median of data.
data = [10,20,30,40,50,60,70,80,90]
data.sort()   # data arranged to ascending order
N = len(data)  # Lenth of data
if N % 2 == 0:  # 
    median = (data[N//2]+data[N//2-1])/2
else:
    median = data[N//2]
print("Median_val =",median)
```

***Find Median with uisng library***

```python
import statistics  # import library
data = [10,20,30,40,50,10,15,25,35]
median = statistics.median(data)
print("Median =",median)
```
