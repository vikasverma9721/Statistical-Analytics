***Mean = The sum of all numbers divided by the count of values.***

***Imputed Data***

```python
# find the imputed data and Original data.
data = [25,20,10,None,15,30,None,24,20,50]
s = 0
count = 0
for x in data:
    if x is not None:
        s += x
        count += 1
mean = s/count
imputed = []
for x in data:
    if x is None:
        imputed.append(mean)
    else:
        imputed.append(x)
print("Mean =",mean)
print("Original Data =",data)
print("Imputed Data =",imputed)
```

***Mean***

```python
# find the mean of dataset without using Library.
dataset = [10,15,24,35,40,12,10]
mean = sum(dataset)/len(dataset)
print("Mean =",mean)
```

***Import Library***

```python
# find the mean dataset with using Library.
# import library
import statistics 
mean_val = statistics.mean(dataset)
print(mean_val)
```
